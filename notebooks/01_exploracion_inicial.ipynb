```python
# Celda 1 — Configuración inicial
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
pd.options.display.float_format = '{:,.2f}'.format

print('✅ Librerías importadas correctamente')
```

```python
# Celda 2 — Cargar el dataset crudo
df_crudo = pd.read_csv('datos/crudos/ventas_colombia_2022_2024.csv', low_memory=False)

print(f'Filas:    {len(df_crudo):,}')
print(f'Columnas: {df_crudo.shape[1]}')
df_crudo.head()
```

```python
# Celda 3 — Diagnóstico de calidad
print('=== VALORES NULOS POR COLUMNA ===')
nulos = df_crudo.isnull().sum()
pct   = (nulos / len(df_crudo) * 100).round(1)
tabla_nulos = pd.DataFrame({'Nulos': nulos, '%': pct})
print(tabla_nulos[tabla_nulos['Nulos'] > 0])

print(f'\n=== DUPLICADOS ===')
print(f'Filas duplicadas: {df_crudo.duplicated().sum():,}')

print(f'\n=== TIPOS DE DATO ===')
print(df_crudo.dtypes)
```

```python
# Celda 4 — Estadísticas descriptivas (columnas numéricas)
df_crudo.describe().round(0)
```

```python
# Celda 5 — Distribución por canal de venta
fig, ax = plt.subplots(figsize=(8, 4))
df_crudo['canal_venta'].value_counts().plot(kind='bar', ax=ax, color='steelblue')
ax.set_title('Distribución por Canal de Venta', fontsize=13, fontweight='bold')
ax.set_xlabel('Canal')
ax.set_ylabel('Número de órdenes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Celda 6 — Distribución por ciudad (top 10)
fig, ax = plt.subplots(figsize=(10, 4))
df_crudo['ciudad'].value_counts().head(10).plot(kind='barh', ax=ax, color='teal')
ax.set_title('Top 10 Ciudades por Número de Órdenes', fontsize=13, fontweight='bold')
ax.set_xlabel('Número de órdenes')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Celda 7 — Distribución del precio unitario (con outliers)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Sin filtro
df_crudo['precio_unitario'].dropna().hist(bins=50, ax=axes[0], color='steelblue')
axes[0].set_title('Precio unitario (distribución completa)')
axes[0].set_xlabel('COP')

# Percentil 1-99 para ver la distribución real
p01 = df_crudo['precio_unitario'].quantile(0.01)
p99 = df_crudo['precio_unitario'].quantile(0.99)
df_crudo['precio_unitario'].clip(p01, p99).dropna().hist(bins=50, ax=axes[1], color='teal')
axes[1].set_title('Precio unitario (percentiles 1-99)')
axes[1].set_xlabel('COP')

plt.tight_layout()
plt.show()
```

```python
# Celda 8 — Cargar dataset limpio (después de ejecutar main.py)
try:
    df_limpio = pd.read_csv('output/ventas_limpio.csv', low_memory=False)
    print(f'Dataset limpio: {len(df_limpio):,} filas × {df_limpio.shape[1]} columnas')
    print(f'Nulos restantes: {df_limpio.isnull().sum().sum()}')
    print(f'Duplicados restantes: {df_limpio.duplicated().sum()}')
    df_limpio.dtypes
except FileNotFoundError:
    print('⚠️  Dataset limpio no encontrado — ejecuta: python main.py')
```
